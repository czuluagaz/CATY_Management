# Process raw data to resample intervals using interpolate
As the dates are irregular, try to use that irregularity to made a better dataset

```python
# Import libraries
import pandas as pd
import os
import sys
import ydata_profiling

sys.path.append("../")
```

```python
# Variables
path = "../data_storage/"
# path to the json file
raw_json = path + "raw_data.json"

df_rawdata = pd.read_json(raw_json)
```

```python
df_rawdata
```

```python
# Function to split the data in 3 different dataset, each one with date T av days + if gaz gaz columns, if water, water columns, if power, power columns
def split_data(df):
    df_gas = df[df.columns[df.columns.str.contains("gaz")]].copy()
    df_water = df[df.columns[df.columns.str.contains("water")]].copy()
    df_power = df[df.columns[df.columns.str.contains("power")]].copy()
    # add the date column to each dataset
    df_gas.loc[:,"date"] = df["date"]
    df_water["date"] = df["date"]
    df_power["date"] = df["date"]
    return df_gas, df_water, df_power

df_gas, df_water, df_power = split_data(df_rawdata)

df_water.info()
```
